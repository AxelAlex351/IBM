**Strings** - en Python los strings son una secuencia de car√°cteres entre dos comillas, pueden ser una o dos comillas ('',""), tambi√©n pueden ser car√°cteres especiales. Se puede asignar un string a una variable, se puede pensar en un string como en una secuencia y cada elemento puede ser accesado: 

image.png


Se puede accesar a cada car√°cter mediante el uso de √≠ndices, empezando por el 0, se puede acceder al 6 o al 14.
Tambi√©n se pueden usar √≠ndices negativos para representar un string. Para esto se usa el comando: **Name[-1]** o similar.

Es posible formar vectores de *n* elementos mediante un conjunto de strings: **Name[0:4]** te devuelve un conjunto de los √≠ndices que forman el string mediante representaci√≥n: **[0123]** en el caso de la imagen superior es la palabra **Mich**.

El comando **len( )** te da el largo del vector de √≠ndices de todo el string.

El comando para concatenar o unir dos strings (pegar uno de lado del otro) se usa mediante un signo +. **Name + 'is the best'** ‚Üí Michael Jackson is the best.
